```python
checkpoint = 'facebook/nllb-200-distilled-600M'
# checkpoint = 'facebook/nllb-200-1.3B'
# checkpoint = 'facebook/nllb-200-3.3B'
# checkpoint = 'facebook/nllb-200-distilled-1.3B'
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline
model = AutoModelForSeq2SeqLM.from_pretrained(checkpoint)
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
```

```python
target_lang = 'urd_Arab'
translation_pipeline = pipeline('translation',
                                model=model,
                                tokenizer=tokenizer,
                                src_lang=input_lang,
                                tgt_lang=target_lang,
                                max_length = 400)
output = translation_pipeline(text)
print(output[0]['translation_text'])
```

```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Translation_dataset/translationDataset.csv")
dataset = []
for ind in df.index:
  dataset.append({"id": ind,
                  "translation":{
                    "en": df['English'][ind],
                    'ur':df['Urdu'][ind]
                  }})

d = pd.DataFrame(dataset)
```

```python
d
```

```python
from datasets import load_dataset, load_metric, Dataset
dataset = Dataset.from_pandas(pd.DataFrame(data=d), split='train')
```

```python
dataset = dataset.train_test_split(test_size = 0.2)
```

```python
dataset['train'][1]
```

```python
from transformers import NllbTokenizerFast

from transformers import AutoModelForSeq2SeqLM, AutoTokenizer, Seq2SeqTrainingArguments
from transformers import DataCollatorForSeq2Seq, Seq2SeqTrainer
```

```python
!pip install accelerate -U
```

```python
tokenizer = NllbTokenizerFast.from_pretrained(
    "facebook/nllb-200-distilled-600M", src_lang="eng_Latn", tgt_lang="urd_Arab"
)

model = AutoModelForSeq2SeqLM.from_pretrained("facebook/nllb-200-distilled-600M")
data_collator = DataCollatorForSeq2Seq(tokenizer=tokenizer, model=model)
```

```python
training_args = Seq2SeqTrainingArguments(
    output_dir="/content",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    weight_decay=0.01,
    save_total_limit=3,
    num_train_epochs=2,
    predict_with_generate=True,
    fp16=True,
)
```

```python
max_length = 128


def preprocess_function(examples):
    inputs = [ex["en"] for ex in examples["translation"]]
    targets = [ex["ur"] for ex in examples["translation"]]
    model_inputs = tokenizer(
        inputs, text_target=targets, max_length=max_length, truncation=True
    )
    return model_inputs
```

```python
tokenized_datasets = dataset.map(
    preprocess_function,
    batched=True,
    remove_columns=dataset['train'].column_names,
)
```

```python
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["test"],
    tokenizer=tokenizer,
    data_collator=data_collator,
)
```

```python
trainer.train()
```
